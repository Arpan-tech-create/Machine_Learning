```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.size']=15
plt.rcParams['figure.figsize']=(15,7)
sns.set_style('darkgrid')
```

```python
raw_data=pd.read_csv('indian_food.csv')
raw_data
```

```python
sns.scatterplot(x=raw_data.prep_time,y=raw_data.cook_time)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
reg.fit(raw_data[['prep_time']],raw_data.cook_time)
```

```python
sns.scatterplot(x=raw_data.prep_time,y=raw_data.cook_time,color='green',s=150)
plt.plot(raw_data.prep_time,reg.predict(raw_data[['prep_time']]),linewidth=7)
plt.legend(['Regression_Line','Points'])
```

```python
prediction=reg.predict(raw_data[['prep_time']])
prediction
```

```python
raw_data['Predicted_cook']=prediction
raw_data
```

```python
new_data=raw_data[['cook_time','Predicted_cook']]
new_data
```

```python
plt.plot(raw_data['cook_time'])
plt.plot(raw_data['Predicted_cook'])
plt.legend(new_data)
```
