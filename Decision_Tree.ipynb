```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.size']=15
plt.rcParams['figure.figsize']=(12,6)
sns.set_style('darkgrid')
```

```python
data=sns.load_dataset('iris')
```

```python
data
```

```python
plt.xlabel('Sepal-Length')
plt.hist(data['sepal_length'],bins=5)
plt.show()
```

```python
plt.xlabel('Petal_Length')
plt.hist(data['petal_length'],color='red')
plt.show()
```

```python
sns.distplot(data['sepal_width'],color='g')
```

# Decision_Tree

```python
#Selecting Target Variables
y=data[['species']]
x=data[['sepal_length']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
dec= DecisionTreeClassifier()
```

```python
dec.fit(x_train,y_train)
```

```python
y_pred=dec.predict(x_test)
```

```python
y_pred
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
#accuracy of model to predict accurate values

(12+10+10)/(12+3+10+6+4+10)
```

```python
from sklearn import tree
tree.plot_tree(dec)
plt.show()
```

# 2nd Model of Decision Tree

```python
x1=data[['sepal_length','petal_length']]
y1=data[['species']]
```

```python
y1_pred=desc.predict(x1_test)
```
